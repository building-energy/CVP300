# Variation of extraterrestrial radiation with time of year

This notebook recreates Equation 1.4.1a and Figure 1.4.1 from the book 'Solar Engineering of Thermal Processes' by Duffie and Beckman.

## Function to calculate G_on - Equation 1.4.1a, Page 9

This function uses several methods from the Python built-in *math* library: https://docs.python.org/3/library/math.html 

```python
import math
def calculate_G_on(n):
    """"Calculates exterterrestrial radiation incident on the plane normal to the radiation
    
    """
    G_sc = 1367
    
    G_on = G_sc * ( 1 + 0.033 * math.cos( math.radians(360 * n / 365) ) ) 
    
    return G_on
    
```

We can check `calculate_G_on` is a function by printing out it's type:

```python
type(calculate_G_on)
```

### Testing the function

We can check the function works by running it for n=1 (1st January) and printing the result

```python
result=calculate_G_on(1)
result
```

## Recreating Figure 1.4.1 from Duffie and Beckman, Page 9

First a list from 1 to 365 is created:

```python
n_list=list(range(1,366))
print(n_list)
```

Next a loop is used to iterate through `n_list` and run the function `calculate_G_on` for each value of n. The results are stored in the variable `results`:

```python
result=[]
for n in n_list:
    result.append(calculate_G_on(n))
print(result)
```

### Plotting the graph

An intial graph of the results is plotted as:

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,figsize=(8,6))
ax.plot(n_list,result)
plt.savefig('Figure1_4_1a.png', bbox_inches='tight')
plt.show()
```

The graph can be improved by including months in the xaxis. This is done by creating a list of the days of a sample year (in this case 2019) and placing it in the variable `days_in_year`. The pandas function date_range is used to create the ist of dates: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html

```python
import pandas as pd
days_in_year=pd.date_range(start=pd.datetime(2019,1,1),freq="1D",periods=365)
print(days_in_year)
```

The dates can now be included in the graph as below. Here the xaxis is formatted so that it displays each month (rather than, say, each day) and it uses the format code '%b' which is the code for the first three letters of each month. 

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

fig, ax = plt.subplots(1,figsize=(8,6))
ax.plot_date(days_in_year,result,'-',xdate=True)

import matplotlib.dates as mdates
months = mdates.MonthLocator()  
months_fmt = mdates.DateFormatter('%b')
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(months_fmt)

plt.savefig('Figure1_4_1b.png', bbox_inches='tight')
plt.show()
```

Finally xaxis and yaxis labels are added to the plot. For the yaxis label, the subscripts and superscipts are included by using dollar signs at the start and end of the string, and then using the Latex commands `_` and `^`.

```python
fig, ax = plt.subplots(1,figsize=(8,6))
ax.plot_date(days_in_year,result,'-',xdate=True)

months = mdates.MonthLocator()  
months_fmt = mdates.DateFormatter('%b')
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(months_fmt)

ax.xaxis.set_label_text('Month')
ax.yaxis.set_label_text('$G_{on}.W/m^2$')

plt.savefig('Figure1_4_1c.png', bbox_inches='tight')
plt.show()
print('Figure 1.4.1 Variation of extraterrestrial solar radiation with time of year.')
```

This plot is now nearly identical to the plot in the textbook.
